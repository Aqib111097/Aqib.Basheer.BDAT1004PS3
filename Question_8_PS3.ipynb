```python
# Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2. Read the CSV file
actor_df = pd.read_csv(r'E:\python\actor_kill_counts.csv')

# Step 3. Sort the data by kill count
actor_df = actor_df.sort_values(by='Count', ascending=True)

# Step 4. Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(actor_df['Actor'], actor_df['Count'], color='skyblue')

# Step 5. Label each bar with the corresponding actor's name
for index, value in enumerate(actor_df['Count']):
    plt.text(value, index, str(value))

# Step 6. Label the axes
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')

# Show plot
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
