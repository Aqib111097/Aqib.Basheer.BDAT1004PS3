```python
# Step 1 Import 
import pandas as pd
import matplotlib.pyplot as plt


# Step 3 Load 
data = pd.read_csv(r'E:\python\us-marriages-divorces-1867-2014.csv')

# Step 4. Filter data for the required years (1867 - 2014)
filtered_data = data[(data['Year'] >= 1867) & (data['Year'] <= 2014)]

# Step 5. Create a line plot
plt.figure(figsize=(10, 6))

# Plot marriages per capita
plt.plot(filtered_data['Year'], filtered_data['Marriages_per_1000'], label='Marriages per capita')

# Plot divorces per capita
plt.plot(filtered_data['Year'], filtered_data['Divorces_per_1000'], label='Divorces per capita')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita Rate')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
