```python

import pandas as pd
import matplotlib.pyplot as plt

roman_emperors_df = pd.read_csv(r'E:\python\roman-emperor-reigns.csv')


num_assassinated = roman_emperors_df[roman_emperors_df['Cause_of_Death'] == 'Assassinated'].shape[0]

# Step 4. Count the number of emperors who were not assassinated
num_not_assassinated = roman_emperors_df.shape[0] - num_assassinated

# Step 5. Create a pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [num_assassinated, num_not_assassinated]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # explode the 1st slice (Assassinated)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.title('Fraction of Roman Emperors Assassinated')
plt.show()
```
