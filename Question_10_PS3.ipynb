```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv(r'E:\python\arcade-revenue-vs-cs-doctorates.csv')


filtered_data = data[(data['Year'] >= 2000) & (data['Year'] <= 2009)]


plt.figure(figsize=(10, 6))


for year, group in filtered_data.groupby('Year'):
    plt.scatter(group['Computer Science Doctorates Awarded (US)'], group['Total Arcade Revenue (billions)'], label=year)


plt.xlabel('Computer Science Doctorates Awarded (US)')
plt.ylabel('Total Arcade Revenue (billions)')
plt.title('Relationship between Arcade Revenue and Computer Science PhDs (2000-2009)')
plt.legend(title='Year')


plt.grid(True)
plt.tight_layout()
plt.show()
```
